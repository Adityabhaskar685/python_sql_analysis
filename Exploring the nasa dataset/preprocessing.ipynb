```python
# imports
import pandas as pd
import numpy as np

%load_ext lab_black
```

```python
df = pd.read_csv("nasa.csv")
df.head()
```

```python
null_values = df.isna().sum()
null_values
```

```python
df = df.dropna(subset=["Alma_Mater"])

idx = df[(df["Death_Mission"].isna()) & (df["Death_Date"].notnull())].index
df = df.drop(idx)

df["Birth_Date"] = pd.to_datetime(df["Birth_Date"], dayfirst=True)
df["Death_Date"] = pd.to_datetime(df["Death_Date"], dayfirst=True)
```

```python
df.to_csv("astronaut.csv", index=False)
```
